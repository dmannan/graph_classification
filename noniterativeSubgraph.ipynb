### This notebook is to demo the non-iterative subgraph extraction procedure


```python
import mgcpy
import numpy as np
#np.random.seed(8889)
from sklearn.metrics import adjusted_rand_score
from sklearn.cluster import KMeans
import graspy
from graspy.embed import AdjacencySpectralEmbed 
from graspy.simulations import sbm
from graspy.plot import heatmap, pairplot
import pandas as pd

from mgcpy.independence_tests.dcorr import DCorr
from mgcpy.independence_tests.mgc.mgc import MGC  ###yay this works!
```

```python
#1. create simulation graphs: Let's create the data for the 1st class with probability 0.3
 
block1= np.ones((20,20))*0.3
block2 = np.ones((20,180))*0.2
block3 = np.ones((180,20))*0.2
block4 = np.ones((180,180))*0.3
block = np.block([[block1, block2], [block3, block4]])
class1=[]
for i in range(100):
    class1.append(block)
len(class1)
    

```

```python
### Creating class 2, prob class 0.4

blo1 = np.ones((20,20))*0.4
blo2 = np.ones((20,180))*0.2
blo3 = np.ones((180,20))*0.2
blo4 = np.ones((180,180))*0.3

blo = np.block([[blo1,blo2],[blo3,blo4]])

class2=[]
for i in range(100):
    class2.append(blo)
len(class2)


```

#### Finding ASE of the matrices

```python
ase = AdjacencySpectralEmbed()
ase1 = [ase.fit_transform(graphss) for graphss in class1]
ase2 = [ase.fit_transform(graphsss) for graphsss in class2]
```

```python
len(ase1)
len(ase2)
ase1[0].shape
```

```python
print(ase1[0]) ## the first 20 vertices are signal vertices and they are different from the rest of the vertices
print(ase2[0]) ## the first 20 vertices are signal vertices 
```

#### Creating class labels

```python
class_labe0 = np.zeros((200,1))
class_label1  = np.ones((200,1))

```

#### Non-iterative vertex screeing method

```python
mgc = MGC()
for features in ase1:
    p_value, metadata = mgc.p_value(features, class_label1, replication_factor = 100)
    mgc_statistic, test_statistic_metadata = mgc.test_statistic(features, class_label1)
    
    print(p_value)
    print(metadata)
    print(mgc_statistic)
    print(test_statistic_metadata)
```
